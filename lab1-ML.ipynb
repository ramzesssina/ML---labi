# **4. Визуализация распределения переменных**

```python
sns.histplot(df["WeekofPurchase"], kde = True)
plt.title("Кол-во недельных закупок")
plt.xlabel("Недельные закупки")
plt.ylabel("Частота")
plt.show()
```

```python
sns.countplot(x = "PriceCH", data=df)
plt.title("Распределение PriceCH")
plt.xlabel("PriceCH")
plt.ylabel("Количество")
plt.show()
```

```python
sns.boxplot(x = df["LoyalCH"])
plt.title("BOXPLOT LoyalCH")
plt.xlabel("LoyalCH")
plt.show()
```

1) Недельные закупки построенные адекватно, ничего лишнего, сначала закупка шла на уменьшение, потом закупка немного резко взлетела.
2) Со вторым графиком все по-другому, даже нет среднего значения, средний ценик в районе 1.69 - 1.79. Дальше идет резкий взлет, значит основные транзакции проходят по этому ценику - 1.86. Также заметен и 1.99 почти на уровне с 1.86. Значит это основные цены транзакций на рынке
3) Медиана сосредоточена на 0.6, между 0.3 и 0.9 - вполне нормально, график без выбросов, но очень большое отклонение от графика.
